### Q1: Summarize the background of Myki data release:

To encourage research and innovation, Public Transport Victoria published a sizable Myki travel dataset in 2018. Touch-on and touch-off records gathered between 2015 and 2018 were included in the dataset. Due to the removal of direct identifiers such names, Myki card numbers, and account information, the data was referred to as "de-identified" prior to release. Instead, a random identity was given to each card, allowing trips to remain connected over time.

Nevertheless, the dataset still included comprehensive details about people's travel times and locations even after direct identifiers were eliminated. Individual travel patterns could be tracked over extended periods of time since every card had a constant ID throughout the dataset. This meant that rather than being completely anonymized, the data was essentially pseudonymized.
When this dataset was connected to publicly accessible external data, re-identification became feasible. For instance, people frequently post about their daily schedules on social media, including when they commute or visit to destinations like companies or colleges. Researchers were able to correlate distinctive travel patterns to particular people by combining this type of data with the Myki data.

The data release posed a genuine risk to privacy, according to the Office of the Victorian Information Commissioner (OVIC). The research revealed flaws in the de-identification procedure and demonstrated that the ease with which data may be connected to other sources was not sufficiently considered. OVIC concluded that, in terms of Victorian privacy law, the dataset might still be regarded as personal information.

### Q2: Compare how the General Data Protection Regulation (GDPR) and the Privacy Act 1988 refer to re-identification:

While both the Australian Privacy Act 1988 and the GDPR address personal data, they handle re-identification risk in different ways. Anonymized and pseudonymized data are clearly distinguished under the GDPR. Only when re-identification is not plausible, taking into consideration current technology, cost, time, and potential advancements, is data deemed anonymous. Data is considered personal data and is protected by the law if it may still be connected to specific people, even in an indirect way.
Re-identification is not covered in as much depth by the Privacy Act of 1988. Rather, it defines personal information as details about a person who may be "reasonably identified." This greatly depends on context and allows for greater space for interpretation.

One significant distinction is that GDPR acknowledges the dangers of data linking and contemporary analysis methods more explicitly. Although the goal of both legislation is to safeguard individuals, GDPR offers more precise guidelines regarding what constitutes truly anonymized data. On the other hand, even while re-identification is practically feasible, the Australian approach may leave gaps where datasets are treated as anonymous.

### Q3: Explain how the Victorian Privacy and Data Protection Act 2014 (PDP Act) define personal information as information about an individual whose identity “can reasonably be ascertained”. Discuss how this implicitly includes re-identification through data linkage. Argue that determining compliance with the PDP Act requires evaluating how easily a member of the public could link the released data with external information to re-identify individuals.

Personal information is defined by the Victorian Privacy and Data Protection Act 2014 as details about a person whose identity "can reasonably be ascertained." Re-identification by data linkage is implicitly included in this definition, which makes it significant.

By merging the Myki dataset with outside data, people may still be identified even if names and card numbers were eliminated. This implies that the information might still be considered personal.

Therefore, determining whether identifiers have been removed is not enough to determine compliance with the PDP Act. Data custodians must consider how simple it would be for a member of the public to identify an individual by connecting the published data with additional information. Even when de-identification procedures are employed, the Myki example demonstrates that disregarding this larger context might result in privacy violations.

### Q4: Research at least one other anonymized dataset that was later re-identified. Describe how linkage occurred, what auxiliary datasets were used, and what consequences followed.

The Netflix Prize dataset, which was made public in 2006, is a famous illustration of re-identification. To facilitate a recommendation competition, Netflix released anonymized movie rating data. Random IDs were used in place of usernames.

However, researchers then shown that by matching Netflix ratings with publicly accessible IMDb ratings, people might be re-identified. Numerous individuals rated the same films on both sites, and people could be uniquely identified using the combination of rating values and timestamps.

Some users' viewing history, including potentially sensitive preferences, were thereby made public. Netflix was sued as a result, and further data leaks were halted. This instance amply illustrates how behavioral data can continue to be identifying even in the absence of conventional identifiers.

### Q5: Recommendations for Future Myki Data Releases - Propose mechanisms that data custodians can use to determine whether re-identification is possible and how likely it is (e.g., re-identification testing, privacy risk scoring, or simulation of linkage scenarios).

Several actions could be implemented to lower the risk of re-identification in further Myki data releases.

Before releasing data, data custodians should first perform re-identification testing. To determine how readily people could be recognized, this entails modeling genuine linking assaults using publicly accessible data.

Second, the degree to which each person's travel habits are distinct might be assessed using privacy risk scoring. Highly unique records may be eliminated or generalized.

Third, methods for minimizing data should be used. This can involve limiting the amount of time a single card can be recorded, classifying places into larger zones, or decreasing time precision.

Lastly, sensitive datasets could be shared through controlled research environments where access is regulated and restricted, as opposed to fully open public releases.
